```python

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib.ticker import PercentFormatter
from sqlalchemy import create_engine

engine = create_engine('mysql+mysqlconnector://root:882489@localhost:3306/Project2')

Connection = engine.connect()
query = '''SELECT * FROM Prod_Info'''
Data  = pd.read_sql_query(query, engine)

fig, axes = plt.subplots(nrows=2, ncols=2)

# 1. Product Count by Branding
Data["Product_count"] = Data.groupby('Branding')['Branding'].transform('count')
Brand_Prod = (Data.loc[:,['Branding','Product_count']]).drop_duplicates().sort_values(by='Product_count',ascending=False).reset_index().drop(columns = 'index')

Brand_Prod = Brand_Prod[Brand_Prod.Branding != "Null"]
Brand_Prod.plot(ax=axes[0,0],x='Branding', y='Product_count',  kind='barh')
axes[0,0].set_xlabel('Number of Products(times)')
axes[0,0].set_ylabel('Brands')
axes[0,0].set_title('Top 20 brands with the highest number of products ')

# 2. Price distribution of Products(the most common price levels)
bins =[]
i = 0 
for i in range(0,8500,400):
    bins.append(i)

Data_Price = Data[Data.Price != 0]
Data_Price.plot(ax=axes[0,1], y='Price',bins=bins,  kind='hist',color='#86bf91', zorder=2, rwidth=0.9)

axes[0,1].set_ylabel('Frequency of Price levels (times)')
axes[0,1].set_xlabel('Price($)')
axes[0,1].set_title('Price distribution of Products')

# 3. Price distribution of Products for every brands
Brand_Price = Data[Data.Branding != "Null"]
Brand_Price.plot(ax=axes[1,0],x='Branding', y='Price', kind='scatter',figsize=(20,20),c='g',rot = 90,s=20, alpha=0.2)
axes[1,0].set_ylabel('Frequency of Price levels (times)')
axes[1,0].set_xlabel('Brands')
axes[1,0].set_title('Price distribution of Products for every brands')


# 4. Price distribution of Ratings
Data_Rating = Data[Data.Rating != "Null"]
Data_Rating.plot(ax=axes[1,1],x='Rating', y='Price', kind='scatter',cmap = 'coolwarm',s=20, alpha=0.3)
sns.regplot(ax=axes[1,1],x=Data['Rating'], y= Data['Price'],label='low')
axes[1,1].set_ylabel('Frequency of Price levels (times)')
axes[1,1].set_xlabel('Rating')
axes[1,1].set_title('Price distribution of Rating')
```
